Euler method to solve two ODEs:
- $y_1'=-y_1$
- $y_2'=-100(y_2 - sin(t)) + cos(t)$

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos

# Define the derivative functions
def f1(y1, t):
    return -y1

def f2(y2, t):
    return -100*(y2 - sin(t)) + cos(t)

# Euler method solver for a system of two ODEs
def euler_system(f1, f2, y1_0, y2_0, h, t_end):
    y1 = [y1_0]
    y2 = [y2_0]
    t = np.arange(0, t_end + h, h)
    for i in range(1, len(t)):
        y1.append(y1[-1] + h * f1(y1[-1], t[i-1]))
        y2.append(y2[-1] + h * f2(y2[-1], t[i-1]))
    return t, y1, y2
```

```python
# Initial conditions
y1_0 = 1
y2_0 = 2
h = 0.01
t_end = 1

t, y1, y2 = euler_system(f1, f2, y1_0, y2_0, h, t_end)

fig, ax = plt.subplots(1,2, figsize=(5,3),dpi=150, sharey=True)
ax[0].plot(t, y1, label='y1')
ax[0].plot(t, y2, label='y2')
ax[0].legend()
ax[0].set_xlabel('t')
ax[0].set_ylabel('y')
ax[0].set_title('h=.01')
ax[0].set_yticks(np.arange(0,2.5,.5))

h = 0.05

t, y1, y2 = euler_system(f1, f2, y1_0, y2_0, h, t_end)

ax[1].plot(t, y1, label='y1')
ax[1].plot(t, y2, label='y2')
ax[1].legend()
ax[1].set_xlabel('t')
ax[1].set_title('h=.05')
ax[1].set_ylim(0,2)

plt.show()
```

#### **Discussion**  

For the smaller step size, the solution for y1 seems reasonable. This indicates that the Forward Euler can handle this type of equation with the given step size. However, there seems to be an issue with the numerical solution and the given IC for y2. However, the issue is not severe, and it seems to "recover" quickly to a sensible solution.

For the larger step size, the numerical solution for y2 is erratic and unstable. This is due to error magnification, and indicates that the method is less able to handle a complicated equation with a larger step size. Clearly, the choice of h is very important to the success of this method.
