# Problem 1

## Part a)

```python
def f1(y,t):
    return -1e6 * y

t1, y1 = forward_euler(f1, y_0=1, t_end=1e-6, h=1e-7)

plt.plot(t1,y1)
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

## Part b)

```python
def f2(y,t):
    return (-1e6 * (y - t**2) + t)

fig, ax = plt.subplots(1,2,figsize=(12,5))

for h in [1e-6,1e-5, 1e-4]:
    t2, y2 = forward_euler(f2, y_0=1, t_end=1, h=h)
    ax[0].plot(t2,y2, label=f'h={h}')
    ax[1].plot(t2,y2, label=f'h={h}')

x = np.arange(stop=1,step=1e-6)
ax[0].plot(x,[(1/1000000000000 + 999999999999*np.exp(-1000000*z)/1000000000000 - z/1000000 + z**2) for z in x], label='True Soln', ls='dashdot')

ax[0].set_ylim(0,1)
ax[0].set_xlim(0,1)
ax[0].legend()

ax[1].set_ylim(0,1)
ax[1].set_xlim(0,5e-6)
ax[1].legend()

plt.show()
```

## Part c

```python
def f3(y,t):
    return -1e6 * (y - sin(1e6*t)) + cos(1e6*t)

for h in [1e-7, 5e-7, 1e-6]:
    t3,y3 = forward_euler(f3,y_0=0, t_end=1, h=h)
    plt.plot(t3,y3, label=f'h={h}')

plt.legend()
plt.xlim(0,6e-6)
plt.show()
```
